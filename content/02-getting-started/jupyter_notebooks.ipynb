```python
# Task 01: Employee Salary Calculator

basic_salary = float(input("Enter basic salary: "))
bonus = float(input("Enter bonus amount: "))
tax_percentage = float(input("Enter tax percentage: "))

gross_salary = basic_salary + bonus
tax_amount = (tax_percentage / 100) * gross_salary
net_salary = gross_salary - tax_amount

print("Net Monthly Salary:", net_salary)
```

```python
# Task 02: Internship Eligibility Checker

cgpa = float(input("Enter CGPA: "))
year = int(input("Enter current year of study: "))
if cgpa >= 7.0 and year >= 3:
    print("Eligible for Internship")
else:
    print("Not Eligible for Internship")
```

```python
# Task 03: Login Authentication System

stored_username = "admin"
stored_password = "9876"
username = input("Enter username: ")
password = input("Enter password: ")
if username == stored_username and password == stored_password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
# Task 04: Shopping Discount Engine

total_amount = float(input("Enter total shopping amount: "))
discount = 15 if total_amount > 1000 else 10
total_amount -= (discount / 100) * total_amount
print("Amount after discount:", total_amount)
```

```python
# Task 05: Course Enrollment Validator

enrolled_courses = ["python", "java", "data science"]
course_name = input("Enter course name: ").lower()
if course_name in enrolled_courses:
    print("Student is enrolled in the course")
else:
    print("Student is not enrolled in the course")
```

```python
# Task 06: Role Verification System
admin_role = {"role": "admin"}
user_role = admin_role
if user_role is admin_role:
    print("Admin access granted")
else:
    print("Access denied")
```

```python
# Task 07: Even-Odd Transaction Checker

transaction_id = int(input("Enter transaction ID: "))

if transaction_id % 2 == 0:
    print("Transaction ID is Even")
else:
    print("Transaction ID is Odd")
```

```python
# Mini Project: Performance Rating System

score = int(input("Enter employee performance score (0-100): "))

if score >= 90:
    rating = "Excellent"
elif score >= 75:
    rating = "Very Good"
elif score >= 60:
    rating = "Good"
elif score >= 40:
    rating = "Average"
else:
    rating = "Poor"

print("Performance Rating:", rating)
```
