```python
def calculator_grade(marks):
    total = sum(marks)
    average = total / len(marks)

    if average >= 80:
        return "A"
    elif average >= 60:
        return "B"
    elif average >= 40:
        return "C"
    else:
        return "Fail"
marks_list = [75, 32, 78, 80, 50]
grade = calculator_grade(marks_list)
average = sum(marks_list) / len(marks_list)
print("Grade:", grade)
print("Average Marks:", average)

```

```python
def withdraw(balance, amount):
    if amount <= balance and amount % 100 == 0:
        balance -= amount
        return balance
    else:
        return "Invalid withdrawal amount or insufficient balance"
print(withdraw(5000, 1200))
```

```python
def password_validator(password):
    if len(password) < 8:
        return "Password too short"

    has_digit = False
    has_upper = False

    for char in password:
        if char.isdigit():
            has_digit = True
        if char.isupper():
            has_upper = True

    if has_digit and has_upper:
        return "Valid Password"
    else:
        return "Invalid Password"
print(password_validator("Python123"))
```

```python
def calculate_salary(basic_salary):
    hra = basic_salary * 0.90
    pf = basic_salary * 0.06
    net_salary = basic_salary + hra - pf
    return net_salary
print(calculate_salary(30000))
```

```python
def file_data_counter(filename):
    lines = 0
    words = 0
    characters = 0
    with open(filename, 'r') as file:
        for line in file:
            lines += 1
            words += len(line.split())
            characters += len(line)
    return lines, words, characters

# Create a sample file for demonstration
with open('sample.txt', 'w') as f:
    f.write('This is the first line.\n')
    f.write('This is the second line with more words.\n')
    f.write('Third line.')

result = file_data_counter("sample.txt")
print("Lines:", result[0])
print("Words:", result[1])
print("Characters:", result[2])
```
